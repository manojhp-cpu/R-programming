# Quick Sort

```python
quick_sort <- function(arr){
  if (length(arr) <= 1)
    return (arr)

  pivot <- arr[length(arr) %/% 2]
  left <- arr[arr < pivot]
  middle <- arr[arr == pivot]
  right <- arr[arr > pivot]

  return(c(quick_sort(left),middle,quick_sort(right)))
}


vect = c(2,5,3,6,8,4,1,3,10)
print("Unsorted Vector" )
print(vect)
sorted_vector <- quick_sort(vect)
print("sortedvector")
print(sorted_vector)



Output:

[1] "Unsorted Vector"
[1]  2  5  3  6  8  4  1  3 10

[1] "sortedvector"
[1]  1  2  3  3  4  5  6  8 10
```


# Binary Search Tree (BST) Implementation in R

``` python
#Define the structure for a BST node
bst_node <- function(key) {
  return(list(
    key = key,
    left = NULL,
    right = NULL
  ))
}

# Function to insert a key into the BST
insert <- function(root, key) {
  if (is.null(root)) {
    return(bst_node(key))
  }

  if (key < root$key) {
    root$left <- insert(root$left, key)
  } else if (key > root$key) {
    root$right <- insert(root$right, key)
  }

  return(root)
}

# Function to perform an in-order traversal of the BST
in_order_traversal <- function(root) {
  if (!is.null(root)) {
    in_order_traversal(root$left)
    cat(root$key, " ")
    in_order_traversal(root$right)
  }
}

# Example usage:
bst <- NULL
keys <- c(5, 3, 8, 1, 9, 2)

for (key in keys) {
  bst <- insert(bst, key)
}

cat("In-order traversal of BST:", "\n")
in_order_traversal(bst)
```
